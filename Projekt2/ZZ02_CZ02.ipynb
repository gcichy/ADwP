Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# Całość danych
# YOUR CODE HERE
#recession_data_1 = recession_data_1.set_index('Country Code')
interval = pd.IntervalIndex.from_arrays([-20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30],[ -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35])
recession_table_1 = pd.DataFrame(columns = interval)
recession_table_1
recession_table_1['Year'] = ['2006', '2007','2008','2009','2010','2011']
recession_table_1 = recession_table_1.set_index('Year')

for x, y in recession_data_1[['2006', '2007','2008','2009','2010','2011']].items():
    d = recession_data_1[x].groupby(pd.cut(y,bins=interval)).count()
    recession_table_1.loc[x] = d/sum(d)

with open('recession_table_1.pkl', 'wb') as handle:
    pickle.dump(recession_table_1, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
dfi.export(recession_table_1.style.format("{:.2f}"), 'calosc.png')
recession_table_1.style.format("{:.2f}")
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
interval = pd.IntervalIndex.from_arrays([-20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30],[ -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35])
index = pd.MultiIndex.from_product([recession_data_1['Income Group'].unique(), ['2006', '2007','2008','2009','2010','2011']])

recession_table_3 = pd.DataFrame(columns = interval, index = index)

for r in recession_data_1['Income Group'].unique():
    for x, y in recession_data_1[['2006', '2007','2008','2009','2010','2011']].items():
        d = recession_data_1[recession_data_1['Income Group']==r][x].groupby(pd.cut(y,bins=interval)).count()
        recession_table_3.loc[r,x] = d/sum(d)
        
#nie dziala:
##dfi.export(recession_table_3.style.format("{:.2f}"), 'rec3.png')
recession_table_3.style.format("{:.2f}")
```

#### c) Dyskusja wyników

Zastanów się jakie są mocne i słabe strony poszczególnych typów wykresów wykonanych w tym zadaniu. W jakich przypadkach wykorzystanie konkretnych typów wykresów do wizualizacji może sprawiać problemy?

Histplot:

Mocne strony:
- Pozwala w sposób jakościowy, ale też ilościowy (określając częstotliwość z jaką dana zmienna przyjmuje wartości z danego bina) analizować rozkład punktów.
- Oprócz tradycyjnego histogramu umożliwia tworzenie diagramów częstości rysowanych linią ciągłą.

Słabe strony:
- Nieczytelny w przypadku nakładania na siebie nawet 3 wykresów jednocześnie
- Nie uwzględnia statystyk np. mediana, jak na przykład boxplot


Boxplot:

Mocne strony:
- Pozwala na łatwe porównywanie rozkładów między grupami danych.
- Pozwala na łatwe wykrycie wartości odstających i skrajnych wartości.
- Przydatny w analizie wielu zmiennych jednocześnie.

Słabe strony:
- Nie ukazuje dokładnej struktury rozkładu, a jedynie jego podstawowe statystyki.
- Nie jest w stanie pokazać szczegółowo rozkładu danych o skomplikowanej strukturze.
- W przypadku skośnych rozkładów może być trudny do zrozumienia.



Violinplot:

Mocne strony:
- W odróżnieniu od boxplota, pokazuje przybliżony kształt rozkładu.
- Niesie pełniejszą informację o tym jak dane rozkładają się w przestrzeni (boxplot tylko Q1, Q2, Q3).


Słabe strony:
- Nie pokazuje ile jest wartości odstających.
- Moim zdaniem jest mniej intuicyjny w rozumieniu.


Swarmplot:

Mocne strony:
- Przekazuje więcej informacji o wyglądzie rozkładu niż violinplot, bo umieszczone są na nim punkty z rozkładu
- Dobry do analizy niewielkich zbiorów danych, gdyż łatwo go wtedy analizować

Słabe strony:
- W przypadku dużych zbiorów danych może być trudny do odczytania.
- Nie pokazuje statystyk jak violinplot czy boxplot
- Niepraktyczny w przypadku dużych zbiorów danych, ponieważ może prowadzić do nakładania się punktów i zasłaniania wartości


Stripplot:

Mocne strony:
- Pozwala jakościowo ocenić jak dużo punktów z rozkładu przyjmuje daną wartość.
- Dobry szczególnie do analizy niewielkich zbiorów danych.
- Łatwy do odczytania przez osoby niezaznajomione z analizą statystyczną.

Słabe strony:
- W przypadku dużych zbiorów danych może być trudny do odczytania, gdyż punkty będą się na siebie nakładać, w efekcie czego możemy otrzymać jeden, gruby pas
- Nie pokazuje podstawowych statystyk.
