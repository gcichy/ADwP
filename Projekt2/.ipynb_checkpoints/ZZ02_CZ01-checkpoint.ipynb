```python
# YOUR CODE HERE
# dane1 = pd.read_csv('raw_data_1.csv', na_values = '..')
# dane2 = pd.read_csv('raw_data_2.csv', sep = ';')
# dane3 = pd.read_excel('raw_data_3.xlsx',skipfooter=2,header=5, sheet_name='Country Analytical History')


# colN = { str(y) + ' [YR' + str(y) + ']' : str(y) for y in range(2006,2012)}
# colN2 = ['Country Code', 'Country Name'] + [str(y)  for y in range(2006,2012)]
# dane1 = dane1.rename(columns = colN)
# dane1 = dane1[[*colN2]]
# dane1 = dane1.dropna() 

# dane2 = dane2.rename(columns={'ISO-alpha3 Code' : 'Country Code'})
# dane2 = dane2[['Country Code', 'Region Name']]
# dane2 = dane2.dropna()

# dane3 = dane3.rename(columns={'Unnamed: 0' : 'Country Code',2008:'Income Group'})
# dane3 = dane3[['Country Code', 'Income Group']]
# recession_data_1 = dane1.join(dane2, on = 'Country Code', how = 'inner')
#dane2 = dane2[['Country Code','Region Name']]
#dane3 = dane3[['Country Code','Income Group']]

#recession_data_1 = pd.DataFrame(dane1)
#recession_data_1 = pd.merge(recession_data_1, dane2, on='Country Code',how='left')
#recession_data_1 = pd.merge(recession_data_1, dane3, on='Country Code',how='left')
#print(recession_data_1.columns)

# YOUR CODE HERE

dane1 = pd.read_csv('raw_data_1.csv',na_values="..")
dane2 = pd.read_csv('raw_data_2.csv', sep = ';')
dane3 = pd.read_excel('raw_data_3.xlsx',skipfooter=2,header=5, sheet_name='Country Analytical History',na_values="..")

dane2 = dane2.rename(columns={'ISO-alpha3 Code' : 'Country Code'})
dane3 = dane3.rename(columns={'Unnamed: 0' : 'Country Code',2008:'Income Group'})

dane2 = dane2[['Country Code','Region Name']]
dane3 = dane3[['Country Code','Income Group']]
recession_data_1 = pd.DataFrame(dane1)
recession_data_1 = pd.merge(recession_data_1, dane2, on='Country Code',how='inner')
recession_data_1 = pd.merge(recession_data_1, dane3, on='Country Code',how='inner')
recession_data_1 = recession_data_1 [['Country Code','Country Name','Region Name','Income Group','2006 [YR2006]', '2007 [YR2007]','2008 [YR2008]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]']]
recession_data_1 = recession_data_1.set_index('Country Code')

recession_data_1 = recession_data_1.rename(columns={'2006 [YR2006]':'2006', '2007 [YR2007]':'2007','2008 [YR2008]':'2008', '2009 [YR2009]':'2009', '2010 [YR2010]':'2010', '2011 [YR2011]':'2011'})
recession_data_1['Income Group'] = recession_data_1['Income Group'].replace({'L':'Low','H':'High','LM':'Lower middle','UM':'Upper middle'})
recession_data_1=recession_data_1.dropna()
recession_data_1 = recession_data_1[ (recession_data_1['Country Name'] !='Libya')]
recession_data_1.to_csv('recession_data_1.csv')
recession_data_1.style.hide(subset=recession_data_1.index[5:-5]).format({"2006":"{:.0f}","2007":"{:.2f}","2008":"{:.2f}","2009":"{:.2f}","2010":"{:.2f}","2011":"{:.2f}"})
#raise NotImplementedError()
```
