### Zadnie 2: Prezentacja danych w formie tabelarycznej [10 pkt]

Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

Indeks:
 - Year (wszystkie lata)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Druga tablica (`recession_table_2`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego z rejonów geograficznych.

Indeks:
 - 1 poziomu: Region Name
 - 2 poziomu: Year (wszystkie lata dla każdego z rejonów)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Trzecia tablica (`recession_table_3`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego stopnia zamożności.

Indeks:
 - 1 poziomu: Income Group
 - 2 poziomu: Year (wszystkie lata dla każdej z grup)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex
 
Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# Całość danych
# YOUR CODE HERE
index_col = pd.interval_range(start=-20, end=35, freq=5)
recession_table_1 = pd.DataFrame(columns=index_col)
index_year = ['2006', '2007','2008','2009','2010','2011']
recession_table_1['Year'] = index_year
recession_table_1=recession_table_1.set_index('Year')

for i, value in recession_data_1[index_year].items():
    temp = recession_data_1[i].groupby(pd.cut(value, bins=index_col)).count()
    recession_table_1.loc[i] = temp/sum(temp)
    
recession_table_1.style.format('{:.2f}')
```

```python
# Podział na rejony
# YOUR CODE HERE
regions = recession_data_1['Region Name'].unique()
index_col = pd.interval_range(start=-20, end=35, freq=5)
index_year = ['2006', '2007','2008','2009','2010','2011']
idx = pd.MultiIndex.from_product([regions, index_year])
recession_table_2 = pd.DataFrame(index=idx, columns=index_col)

for j in regions:
    for i, value in recession_data_1[index_year].items():
        temp = recession_data_1[recession_data_1['Region Name']==j][i].groupby(pd.cut(value, bins=index_col)).count()
        recession_table_2.loc[j,i] = temp/sum(temp)
    
recession_table_2.style.format('{:.2f}')
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
inc_group = recession_data_1['Income Group'].unique()
index_col = pd.interval_range(start=-20, end=35, freq=5)
index_year = ['2006', '2007','2008','2009','2010','2011']
idx = pd.MultiIndex.from_product([inc_group, index_year])
recession_table_3 = pd.DataFrame(index=idx, columns=index_col)

for j in inc_group:
    for i, value in recession_data_1[index_year].items():
        temp = recession_data_1[recession_data_1['Income Group']==j][i].groupby(pd.cut(value, bins=index_col)).count()
        recession_table_3.loc[j,i] = temp/sum(temp)
    
recession_table_3.style.format('{:.2f}')
```

Uwaga! Zadanie należy wykonać w oparciu o wczytane do notatnika szeregi szczegółowe, a nie w oparciu o szeregi rozdzielcze częstości przygotowane w zadaniu 2.

```python
# Boxplot
# YOUR CODE HERE
fig, ax = plt.subplots(3,1, figsize=[20,20])

sns.boxplot(melted_data, x='Year', y='value', color='pink', ax=ax[0], zorder=1,showmeans=True, meanprops={"marker":"o", "markerfacecolor":'black'})
ax[0].axhline(0,color='red',zorder=0)
ax[0].set_title("All data", fontsize=17)
ax[0].set_ylabel('Change of GDP',fontsize=10)

sns.boxplot(melted_data, x='Year', y='value', hue="Income Group", ax=ax[1], zorder=1, palette="coolwarm", showmeans=True, meanprops={"marker":"o", "markerfacecolor":'black'})
ax[1].axhline(0,color='red',zorder=0)
ax[1].set_title("Income group", fontsize=17)
ax[1].set_ylabel('Change of GDP',fontsize=10)

sns.boxplot(melted_data, x='Year', y='value', hue="Region Name", ax=ax[2], zorder=1, palette=["gold", "royalblue", "dimgrey", "limegreen", "tomato"], showmeans=True, meanprops={"marker":"o", "markerfacecolor":'black'})
ax[2].axhline(0,color='red',zorder=0)
ax[2].set_title("Region name", fontsize=17)
ax[2].set_ylabel('Change of GDP',fontsize=10)

fig.suptitle("Boxplots of GDP for particular year, depending on geographical region or income group", fontsize=20)
plt.subplots_adjust(wspace=0.4)
fig.set_facecolor('white')
```

```python
# Violinplot
# YOUR CODE HERE
fig, ax = plt.subplots(3,1, figsize=[20,20])

sns.violinplot(melted_data, x='Year', y='value', color='pink', ax=ax[0], zorder=1)
ax[0].axhline(0,color='red',zorder=0)
ax[0].set_title("All data", fontsize=17)
ax[0].set_ylabel('Change of GDP',fontsize=10)

sns.violinplot(melted_data, x='Year', y='value', hue="Income Group", ax=ax[1], zorder=1, palette="coolwarm")
ax[1].axhline(0,color='red',zorder=0)
ax[1].set_title("Income group", fontsize=17)
ax[1].set_ylabel('Change of GDP',fontsize=10)

sns.violinplot(melted_data, x='Year', y='value', hue="Region Name", ax=ax[2], zorder=1, palette=["gold", "royalblue", "dimgrey", "limegreen", "tomato"])
ax[2].axhline(0,color='red',zorder=0)
ax[2].set_title("Region name", fontsize=17)
ax[2].set_ylabel('Change of GDP',fontsize=10)

fig.suptitle("Violinplots of GDP for particular year, depending on geographical region or income group", fontsize=20)
plt.subplots_adjust(wspace=0.4)
fig.set_facecolor('white')
```

```python
# Swarmplot
# YOUR CODE HERE
fig, ax = plt.subplots(3,1, figsize=[20,20])

sns.swarmplot(melted_data, x='Year', y='value', color='pink', ax=ax[0], zorder=1)
ax[0].axhline(0,color='red',zorder=0)
ax[0].set_title("All data", fontsize=17)
ax[0].set_ylabel('Change of GDP',fontsize=10)

sns.swarmplot(melted_data, x='Year', y='value', hue="Income Group", ax=ax[1], zorder=1, palette="coolwarm", dodge=True)
ax[1].axhline(0,color='red',zorder=0)
ax[1].set_title("Income group", fontsize=17)
ax[1].set_ylabel('Change of GDP',fontsize=10)

sns.swarmplot(melted_data, x='Year', y='value', hue="Region Name", ax=ax[2], zorder=1, palette=["gold", "royalblue", "dimgrey", "limegreen", "tomato"], dodge=True)
ax[2].axhline(0,color='red',zorder=0)
ax[2].set_title("Region name", fontsize=17)
ax[2].set_ylabel('Change of GDP',fontsize=10)

fig.suptitle("Swarmplots of GDP for particular year, depending on geographical region or income group", fontsize=20)
plt.subplots_adjust(wspace=0.4)
fig.set_facecolor('white')
```

```python
# Stripplot
# YOUR CODE HERE
fig, ax = plt.subplots(3,1, figsize=[20,20])

sns.stripplot(melted_data, x='Year', y='value', color='pink', ax=ax[0], zorder=1)
ax[0].axhline(0,color='red',zorder=0)
ax[0].set_title("All data", fontsize=17)
ax[0].set_ylabel('Change of GDP',fontsize=10)

sns.stripplot(melted_data, x='Year', y='value', hue="Income Group", ax=ax[1], zorder=1, palette="coolwarm", dodge=True)
ax[1].axhline(0,color='red',zorder=0)
ax[1].set_title("Income group", fontsize=17)
ax[1].set_ylabel('Change of GDP',fontsize=10)

sns.stripplot(melted_data, x='Year', y='value', hue="Region Name", ax=ax[2], zorder=1, palette=["gold", "royalblue", "dimgrey", "limegreen", "tomato"], dodge=True)
ax[2].axhline(0,color='red',zorder=0)
ax[2].set_title("Region name", fontsize=17)
ax[2].set_ylabel('Change of GDP',fontsize=10)

fig.suptitle("Stripplots of GDP for particular year, depending on geographical region or income group", fontsize=20)
plt.subplots_adjust(wspace=0.4)
fig.set_facecolor('white')
```

Boxplot:

Mocne strony:
- Pozwala na łatwe porównywanie rozkładów między grupami danych.
- Pozwala na łatwe wykrycie wartości odstających i skrajnych wartości.
- Przydatny w analizie wielu zmiennych jednocześnie.

Słabe strony:
- Nie ukazuje dokładnej struktury rozkładu, a jedynie jego podstawowe statystyki.
- Nie jest w stanie pokazać szczegółowo rozkładu danych o skomplikowanej strukturze.
- W przypadku skośnych rozkładów może być trudny do zrozumienia.



Violinplot:

Mocne strony:
- Daje pełniejszą informację o rozkładzie danych niż boxplot.
- Pozwala na łatwe porównywanie rozkładów między grupami danych.

Słabe strony:
- W przypadku dużej ilości danych może być trudny do odczytania.
- Nie pokazuje dokładnej liczby wartości odstających.
- Może być trudny do zinterpretowania przez osoby niezaznajomione z tym typem wykresu.



Stripplot:

Mocne strony:
- Pozwala na łatwe zobrazowanie całego zbioru danych.
- Łatwy do odczytania przez osoby niezaznajomione z analizą statystyczną.

Słabe strony:
- W przypadku dużych zbiorów danych może być trudny do odczytania.
- Nie uwzględnia informacji o gęstości punktów.
- Nie pokazuje podstawowych statystyk.



Swarmplot:

Mocne strony:
- Pokazuje całą informację o rozkładzie danych i pozwala na łatwe porównywanie między grupami.
- W przypadku małych zbiorów danych jest łatwy do odczytania.

Słabe strony:
- W przypadku dużych zbiorów danych może być trudny do odczytania.
- Nie uwzględnia informacji o gęstości punktów.


W przypadku dużej ilości danych, stosowanie wykresów stripplot lub swarmplot może sprawiać problemy, ponieważ może to prowadzić do zbyt dużego zagęszczenia punktów, co sprawi, że wykres stanie się nieczytelny. W takim przypadku lepiej stosować wykres boxplot lub violinplot, które pozwolą na łatwe zobrazowanie rozkładu danych w bardziej przystępny sposób.
