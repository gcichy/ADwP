#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE
rt1 = recession_table_1 
columns = ['2006','2007','2008','2009','2010','2011'] 
index = ['mean','std','Q1','Q2','Q3','mode']
descriptive_parameters_2 = pd.DataFrame(columns = columns, index = index)
x = np.array([-17.5, -12.5, -7.5, -2.5, 2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5])

#mean:
for i in range(0,len(columns)):
    N = np.sum(rt1.iloc[i])
    #srednia
    mean =  np.sum(x*rt1.iloc[i,:])/N 
    descriptive_parameters_2.loc['mean',columns[i]] = np.sum(x*rt1.iloc[i,:])/N 
    
    #odch std
    std = np.sum((x-mean)**2*rt1.iloc[i,:])/N
    descriptive_parameters_2.loc['std',columns[i]] = np.sqrt(std)
    
    #moda
    x_d = x - 2.5
    i_d = x[-1]-x[-2]
    n_d = max(rt1.iloc[i,:])
    #indeks wartości max dla danego roku
    ind = np.where(np.array(abs(rt1.iloc[i,:] - max(rt1.iloc[i,:])) < 1e-6))[0][0] 
    mode = x_d[ind] + (rt1.iloc[i,ind] - rt1.iloc[i,ind-1])/(2*rt1.iloc[i,ind] - rt1.iloc[i,ind-1] - rt1.iloc[i,ind+1])*i_d
    descriptive_parameters_2.loc['mode',columns[i]] = mode
    
    #kwantyle
    iq1 = np.where(np.array([np.sum(rt1.iloc[i,0:j]) for j in range(1,len(rt1.iloc[i]))]) > 0.25)[0][0]
    iq2 = np.where(np.array([np.sum(rt1.iloc[i,0:j]) for j in range(1,len(rt1.iloc[i]))]) > 0.50)[0][0]
    iq3 = np.where(np.array([np.sum(rt1.iloc[i,0:j]) for j in range(1,len(rt1.iloc[i]))]) > 0.75)[0][0]
    Q1 = x_d[iq1] + (N/4 - np.sum(rt1.iloc[i,0:iq1]))/rt1.iloc[i,iq1]*i_d
    descriptive_parameters_2.loc['Q1',columns[i]] = Q1
    Q2 = x_d[iq2] + (N/2 - np.sum(rt1.iloc[i,0:iq2]))/rt1.iloc[i,iq2]*i_d
    descriptive_parameters_2.loc['Q2',columns[i]] = Q2
    Q3 = x_d[iq3] + (3*N/4 - np.sum(rt1.iloc[i,0:iq3]))/rt1.iloc[i,iq3]*i_d
    descriptive_parameters_2.loc['Q3',columns[i]] = Q3
descriptive_parameters_2

#.column.get_loc(recession_table_1.iloc[0,:].index[abs(recession_table_1.iloc[0,:] - 0.070707) < 1e-6])
```

```python
from matplotlib.gridspec import GridSpec
from matplotlib.lines import Line2D

def ax_fonts(ax, fsize, title, xlabel, ylabel, ymax = None, yax_col = None, no_rotate = None):
    ax.set_title(title, fontsize = fsize+5)
    ax.set_xlabel(xlabel, fontsize = fsize) 
    if no_rotate == True:
        ax.tick_params(axis='x', labelsize=fsize)
    else:
        ax.tick_params(axis='x', rotation=90, labelsize=fsize)
    
    if yax_col != None:
        ax.set_ylabel(ylabel, fontsize = fsize, color = yax_col) 
        ax.tick_params(axis='y', labelsize=fsize,  color = yax_col, labelcolor = yax_col)
    else:
        ax.set_ylabel(ylabel, fontsize = fsize) 
        ax.tick_params(axis='y', labelsize=fsize)
    if ymax != None:
        ax.set_ylim([0,ymax])
        
    

fig = plt.figure(layout="constrained", figsize=(25,25))

gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1, :])


ax11 = ax1.twinx()
sns.barplot(y = recession_table_1.loc['sum'], x = recession_table_1.columns, ax = ax1, color = 'green')
sns.pointplot(y = recession_table_1.loc['cumulant'], x = recession_table_1.columns, ax = ax11, color = 'r')
ax_fonts(ax1, 25, 'GDP: counts histogram & cumulative frequency' , 'GDP value ranges', 'Percent', 1, 'green')


ax_fonts(ax11, 25, 'GDP: counts histogram & cumulative frequency' , 'GDP value ranges', 'Cumulative frequency', 1, 'red')

pal = sns.color_palette("hls", 6)
years = [str(i) for i in range(2006,2012)]
legend_handles = []
for i in range(len(years)):
    sns.pointplot(y = recession_table_1.loc[years[i]], x = recession_table_1.columns, ax = ax2, color = pal[i])
    legend_handles.append(Line2D([], [], color=pal[i], markersize=50, label=years[i])) 

ax2.legend(legend_handles, years, title='Years', loc='best')
plt.setp(ax2.get_legend().get_lines(), linewidth=8)
plt.setp(ax2.get_legend().get_texts(), fontsize='30')
plt.setp(ax2.get_legend().get_title(), fontsize='30')
ax_fonts(ax2, 25, 'GDP frequency plot by year', 'GDP value ranges', 'Frequency', 0.6, 'black')

dp2 = descriptive_parameters_2
dicts = []
for i in years:
    stats = {'mean': dp2.loc['mean', i],
              'med': dp2.loc['Q2', i],
              'q1': dp2.loc['Q1', i],
              'q3': dp2.loc['Q3', i],
              'whislo': dp2.loc['Q1', i] - 1.5*(dp2.loc['Q3', i]-dp2.loc['Q1', i]),
              'whishi': dp2.loc['Q3', i] + 1.5*(dp2.loc['Q3', i]-dp2.loc['Q1', i])}
    dicts.append(stats)

ax3.bxp(dicts, showfliers=False, showmeans=True,                 \
        meanprops={"marker":"o","markerfacecolor":"white",       \
                   "markeredgecolor":"black","markersize":"15"}, \
        whiskerprops={"color":'black', "linewidth": 4},          \
        capprops={'color':'black', 'linewidth':4},               \
        medianprops={'color':'gold', 'linewidth':4},             \
        boxprops={'color':'gold', 'linewidth':4})
ax3.axhline(y = 0, color = 'r', linestyle = '-')
ax_fonts(ax3, 25, 'Boxplot for descriptive data', 'Years', 'GDP value', no_rotate = True)
        
plt.show
```
